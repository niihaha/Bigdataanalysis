### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Data

```python
data = pd.read_csv('data.csv', low_memory=False)
```

```python
data.info()
```

### Data Cleaning

```python
data.drop(['vin','back_legroom','bed','bed_height','bed_length','cabin','city_fuel_economy','combine_fuel_economy','dealer_zip',
          'description','engine_cylinders','fleet','frame_damaged','has_accidents','highway_fuel_economy','isCab','is_cpo','is_certified',
          'is_oemcpo','latitude','listed_date','listing_id','longitude','power','main_picture_url','major_options','owner_count','salvage',
          'savings_amount','sp_id','sp_name','theft_title','transmission','trimId','trim_name','vehicle_damage_category','wheel_system',
          'wheelbase'], axis=1, inplace=True)
```

```python
data.info()
```

```python
data = data.loc[data['franchise_dealer']==True]
data = data.loc[data['fuel_type']=='Gasoline']
```

```python
data.drop(['franchise_dealer','fuel_type'],axis = 1,inplace = True)
data.dropna(inplace = True)
```

```python
data.info()
```

```python
data.head().T
```

```python
height = data.height.str.split(" ",n = 1, expand = True)
data.height = height[0]
length = data.length.str.split(" ",n = 1, expand = True)
data.length = length[0]
width = data.width.str.split(" ",n = 1, expand = True)
data.width = width[0]
seats = data.maximum_seating.str.split(" ",n = 1, expand = True)
data.maximum_seating = seats[0]
legroom = data.front_legroom.str.split(" ",n = 1, expand = True)
data.front_legroom = legroom[0]
fuel_tank_volume = data.fuel_tank_volume.str.split(" ",n = 1, expand = True)
data.fuel_tank_volume = fuel_tank_volume[0]
torque = data.torque.str.split(" ", expand = True)
data.torque = torque[3]
```

```python
data.head().T
```

```python
data = data[data.mileage<50]
```

```python
cols = ['city','height','length','width','maximum_seating','front_legroom','fuel_tank_volume','exterior_color','engine_type',
        'transmission_display','wheel_system_display']
```

```python
threshold = 10000 # Anything that occurs less than this will be removed.
for col in cols:
    value_counts = data[col].value_counts() # Specific column 
    to_remove = value_counts[value_counts <= threshold].index
    data[col].replace(to_remove, np.nan, inplace=True)
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
columns = ['daysonmarket','engine_displacement','front_legroom','fuel_tank_volume','height','horsepower','length',
          'maximum_seating','mileage','price','seller_rating']
```

```python
data.head().T
```

```python
data.to_csv("cleanedata.csv",index=False)
```
